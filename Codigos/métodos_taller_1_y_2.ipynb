```python
from sympy import symbols, Matrix, Rational, acos, pi, sqrt, simplify, N
from sympy.vector import CoordSys3D, express

def ingresar_vector(nombre):
    print(f"\nIngrese las componentes del vector {nombre} (separadas por espacios):")
    componentes = input().split()
    vector = []
    for comp in componentes:
        if '/' in comp:
            num, den = map(int, comp.split('/'))
            vector.append(Rational(num, den))
        else:
            vector.append(Rational(comp))
    return Matrix(vector)

a1 = ingresar_vector("a")
b1= ingresar_vector("b")
c1 = ingresar_vector("c")

g1 = (a1 + b1 + c1) / 3

print("\n--- Resultados ---")
print(f"Vector a = {a1}")
print(f"Vector b = {b1}")
print(f"Vector c = {c1}")
print("Resultados problema 1:")
print(f"\nCentroide G = (a1 + b1 + c1)/3 = {g1}")
```

```python
from sympy import acos, pi, sqrt, Matrix
from sympy.vector import CoordSys3D, express

S = CoordSys3D('S')  # Sistema de coordenadas
e1, e2, e3 = S.i, S.j, S.k  # Vectores base

# Definición de vectores
a = 1*e1 + 2*e2 + 3*e3
b = 4*e1 + 5*e2 + 6*e3
c = 3*e1 + 2*e2 + 1*e3
d = 6*e1 + 5*e2 + 4*e3

# ===== Funciones auxiliares =====
def vec_to_matrix(v):
    """Convierte un vector de CoordSys3D a matriz."""
    return Matrix([v.coeff(e1), v.coeff(e2), v.coeff(e3)])

def angle_between(v1, v2):
    """Calcula el ángulo entre vectores en grados."""
    dot_p = v1.dot(v2)
    mag_v1 = sqrt(v1.dot(v1))
    mag_v2 = sqrt(v2.dot(v2))
    cos_theta = (dot_p / (mag_v1 * mag_v2)).evalf()
    cos_theta = max(-1, min(1, float(cos_theta)))
    angle_rad = acos(cos_theta)
    return float((angle_rad * 180 / pi).evalf())

# ===== Cálculos principales =====
# (a) Sumas vectoriales (simbólicas)
sum1 = a + b + c + d
sum2 = a + b - c - d
sum3 = a - b + c - d
sum4 = -a + b - c + d

# (b) Ángulos con vectores base
def angle_with_bases(v):
    angles = []
    for base in [e1, e2, e3]:
        dot_product = v.dot(base)
        mag_v = sqrt(v.dot(v))
        if mag_v != 0:
            cos_theta = (dot_product / mag_v).evalf()
            cos_theta = max(-1, min(1, float(cos_theta)))
            angle_rad = acos(cos_theta)
            angle_deg = float((angle_rad * 180 / pi).evalf())
            angles.append(round(angle_deg, 2))
        else:
            angles.append(0.0)
    return angles

angles_a = angle_with_bases(a)
angles_b = angle_with_bases(b)
angles_c = angle_with_bases(c)
angles_d = angle_with_bases(d)

# (c) Magnitudes (numéricas)
mag_a = float(sqrt(a.dot(a)).evalf())
mag_b = float(sqrt(b.dot(b)).evalf())
mag_c = float(sqrt(c.dot(c)).evalf())
mag_d = float(sqrt(d.dot(d)).evalf())

# (d) Ángulos entre vectores
angle_ab = round(angle_between(a, b), 2)
angle_cd = round(angle_between(c, d), 2)

# (e) Proyección de a sobre b
proj_a_on_b = (a.dot(b) / b.dot(b)) * b
proj_a_on_b_num = express(proj_a_on_b.evalf(), S)

# (f) Comprobar coplanaridad
a_mat = vec_to_matrix(a)
b_mat = vec_to_matrix(b)
c_mat = vec_to_matrix(c)
triple_product = float(a_mat.dot(b_mat.cross(c_mat)).evalf())
coplanar = abs(triple_product) < 1e-10

# (g) Producto punto (corregido)
dot_product = (a + b).dot(c + d).evalf()  # Primero evalúa, luego convierte
dot_product_float = float(dot_product)

# (h) Productos cruz
cross_ab = a.cross(b).evalf()
cross_bc = b.cross(c).evalf()
cross_cd = c.cross(d).evalf()

angle_ab_d = round(angle_between(cross_ab, d), 2)
angle_bc_d = round(angle_between(cross_bc, d), 2)
angle_cd_d = round(angle_between(cross_cd, d), 2)

# (i) Triple producto escalar
triple_dot = float(c.dot(a.cross(b)).evalf())

# ===== Resultados =====
print("=== Resultados del Ejercicio 6 ===")
print("\n(a) Sumas vectoriales:")
print(f"1. a + b + c + d = {express(sum1, S)}")
print(f"2. a + b - c - d = {express(sum2, S)}")
print(f"3. a - b + c - d = {express(sum3, S)}")
print(f"4. -a + b - c + d = {express(sum4, S)}")

print("\n(b) Ángulos con vectores base (grados):")
print(f"a: {angles_a}° | b: {angles_b}° | c: {angles_c}° | d: {angles_d}°")

print("\n(c) Magnitudes:")
print(f"|a| = {mag_a:.2f} | |b| = {mag_b:.2f} | |c| = {mag_c:.2f} | |d| = {mag_d:.2f}")

print("\n(d) Ángulos entre vectores:")
print(f"a y b: {angle_ab}° | c y d: {angle_cd}°")

print("\n(e) Proyección de a sobre b:")
print(express(proj_a_on_b, S))

print("\n(f) ¿Son a, b, c coplanares?")
print(f"{'Sí' if coplanar else 'No'} (Producto mixto = {triple_product:.2f})")

print("\n(g) (a + b) · (c + d):")
print(f"{dot_product_float:.2f}")

print("\n(h) Productos cruz:")
print(f"a × b = {express(cross_ab, S)} (Ángulo con d: {angle_ab_d}°)")
print(f"b × c = {express(cross_bc, S)} (Ángulo con d: {angle_bc_d}°)")
print(f"c × d = {express(cross_cd, S)} (Ángulo con d: {angle_cd_d}°)")

print("\n(i) c · (a × b):")
print(f"{triple_dot:.2f}")
```

```python
from sympy import symbols, cos, sin, integrate, pi

theta = symbols('theta')

# Parametrización del círculo
x = cos(theta)
y = sin(theta)

# Campo de fuerza F
F_x = -y / (x**2 + y**2)  # Simplifica a -sin(theta)
F_y = x / (x**2 + y**2)    # Simplifica a cos(theta)

# Diferencial de posición dr
dr_x = -sin(theta)
dr_y = cos(theta)

# Trabajo: F · dr
F_dot_dr = F_x * dr_x + F_y * dr_y

# (a) Trabajo de 0 a π (antihorario)
W_a = integrate(F_dot_dr, (theta, 0, pi))

# (b) Trabajo de 0 a -π (horario)
W_b = integrate(F_dot_dr, (theta, 0, -pi))

print(f"(a) Trabajo (0 a π): {W_a}")
print(f"(b) Trabajo (0 a -π): {W_b}")
```
